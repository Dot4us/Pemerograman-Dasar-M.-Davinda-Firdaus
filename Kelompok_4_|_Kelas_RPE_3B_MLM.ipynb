```python
# 1.Percabangan/Branching

suhu = float(input("Input suhu air pendingin (°C): "))

if suhu > 99.9:
    print("ALERT: SHUTDOWN")
elif 90 < suhu <= 99.9:
    print("STATUS: Warning")
else:
    print("STATUS: Normal")
```

```python
# 3.Looping [for]

vibrasi_mesin = [1.0, 0.9, 2.5, 3.1, 1.8]

for nomor_unit, nilai in enumerate(vibrasi_mesin, start=1):
    if nilai > 2.0:
        print(f"Warning: Unit {nomor_unit} vibrasi tinggi!")
```

```python
# 5. Looping [while]
suhu_saat_ini = 60

while suhu_saat_ini < 85:
    print("Suhu saat ini:", suhu_saat_ini, "°C")
    suhu_saat_ini += 5
```

```python
# 7. Percabangan/Branching

kapasitas_maks = float(input("Masukkan kapasitas maksimum (kW): "))
beban_terpakai = float(input("Masukkan beban terpakai (kW): "))

persentase = (beban_terpakai / kapasitas_maks) * 100

if persentase <= 60:
    print("Status: Beban Ringan")
elif persentase <= 85:
    print("Status: Beban Sedang")
else:
    print("Status: Beban Berat")
```

```python
# 9. Looping + List

suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]

titik = 1
for suhu in suhu_boiler:
    if suhu > 100:
        print("Titik", titik, ": ALERT! Suhu terlalu tinggi", suhu, "°C")
    elif suhu >= 90:
        print("Titik", titik, ": Peringatan! Suhu mendekati batas", suhu, "°C")
    else:
        print("Titik", titik, ": Suhu dalam kondisi normal", suhu, "°C")
    titik += 1
```

```python
# Challege Penambahan Point Nilai

def hitung_daya(arus, tegangan):
    """Menghitung daya (Watt)"""
    return arus * tegangan

def status_baterai(soc):
    """Menentukan status baterai berdasarkan persentase SoC"""
    if soc >= 80:
        return "Penuh"
    elif soc >= 40:
        return "Cukup"
    elif soc >= 20:
        return "Rendah"
    else:
        return "Kritis"

print("=== SISTEM PEMANTAUAN BATERAI KENDARAAN LISTRIK ===")

jumlah_sesi = int(input("Masukkan jumlah sesi pengukuran: "))
total_daya = 0
status_terakhir = ""

for sesi in range(1, jumlah_sesi + 1):
    print(f"\nSesi {sesi}")
    arus = float(input("  Masukkan arus baterai (Ampere): "))
    tegangan = float(input("  Masukkan tegangan baterai (Volt): "))
    soc = float(input("  Masukkan State of Charge (0-100%): "))

    daya = hitung_daya(arus, tegangan)
    total_daya += daya
    status_terakhir = status_baterai(soc)

    print(f"  Daya saat ini: {daya:.2f} Watt")
    print(f"  Status baterai: {status_terakhir}")

# Menampilkan ringkasan
rata_daya = total_daya / jumlah_sesi
print("\n=== RINGKASAN PENGUKURAN ===")
print(f"Total daya terukur: {total_daya:.2f} Watt")
print(f"Rata-rata daya: {rata_daya:.2f} Watt")
print(f"Status terakhir baterai: {status_terakhir}")

# Simulasi loop pemantauan sistem
print("\n=== SIMULASI PEMANTAUAN OTOMATIS ===")
waktu = 1
while waktu <= 3:
    print(f"Monitoring ke-{waktu}: Sistem bekerja stabil...")
    waktu += 1

print("\nPemantauan selesai. Sistem berjalan dengan aman.")
```
