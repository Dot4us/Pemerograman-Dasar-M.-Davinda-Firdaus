```python
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

Perintah delivery_time = orders_df["delivery_date"] - orders_df["order_date"] digunakan untuk menghitung durasi antara tanggal pengiriman dan tanggal pesanan dibuat dengan mengurangkan delivery_date dengan order_date. Hasilnya adalah sebuah timedelta yang menunjukkan selisih waktu antara kedua tanggal tersebut. Kemudian, delivery_time = delivery_time.apply(lambda x: x.total_seconds()) mengkonversi timedelta tersebut menjadi jumlah detik dengan menggunakan fungsi total_seconds(). Terakhir, orders_df["delivery_time"] = round(delivery_time/86400) mengkonversi detik menjadi hari dengan membagi hasilnya dengan 86400 (jumlah detik dalam satu hari) dan membulatkan hasilnya. Ini menghasilkan kolom baru delivery_time yang menunjukkan waktu pengiriman dalam satuan hari.
